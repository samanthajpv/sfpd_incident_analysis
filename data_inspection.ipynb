```python
# import dependencies
import pandas as pd
import numpy as np
```

```python
# data was downloaded on April 17, 2022 from 
# https://data.sfgov.org/Public-Safety/Police-Department-Incident-Reports-2018-to-Present/wg3w-h783
# column information - https://datasf.gitbook.io/datasf-dataset-explainers/sfpd-incident-report-2018-to-present

# load file to pandas dataframe
incident_df = pd.read_csv('Police_Department_Incident_Reports__2018_to_Present.csv')
incident_df.head()
```

```python
# check info
incident_df.info()
```

```python
# check unique values and count of null
for col in incident_df.columns:
    vc = incident_df[col].value_counts()
    null_ct = incident_df[col].isna().sum()
    unique_ct = incident_df[col].nunique()
    print(f'Column: {col}\n'
          f'Count of unique values: {unique_ct}\n'
          f'Count of null: {null_ct}\n'
            f'{vc}\n')
```

```python
# drop columns not needed
incident_df.drop(incident_df.columns[[19,25,26,27,28,29,30,31,32,33]],axis=1,inplace=True)

# replace null values in column 'Filed Online' with False
incident_df['Filed Online'].fillna('False', inplace = True)

# Convert columns to datetime
incident_df['Incident Datetime'] = pd.to_datetime(incident_df['Incident Datetime'])
incident_df['Incident Date'] = pd.to_datetime(incident_df['Incident Date'])
```

```python
# Incident Category
incident_df['Incident Category'].value_counts()
```

```python
incident_df['Incident Category'].replace({"Weapons Carrying Etc":"Weapons Offense",
                                          "Weapons Offence":"Weapons Offense",
                                            "Human Trafficking (A), Commercial Sex Acts":"Human Trafficking",
                                            "Human Trafficking, Commercial Sex Acts":"Human Trafficking",
                                            "Human Trafficking (B), Involuntary Servitude":"Human Trafficking",
                                            "Suspicious":"Suspicious Occ",
                                            "Motor Vehicle Theft?":"Motor Vehicle Theft",
                                            "Rape":"Sex Offense",
                                            "Other Offenses":"Other Miscellaneous",
                                            "Other":"Other Miscellaneous",
                                            "Drug Violation":"Drug Offense",
                                            "Vandalism":"Malicious Mischief",
                                            "Vehicle Misplaced":"Vehicle Impounded/Misplaced",
                                            "Vehicle Impounded":"Vehicle Impounded/Misplaced"}, inplace=True)
```
